```python
# 일단 카테고리별로 선택해서 들어가야 한다고 생각 했으나, 안에 데이터가 display=none으로 되어 있음
# 그러므로 그냥 한 페이지에서 다 뽑아낼 수 있음.

def category():

    driver = webdriver.Chrome()
    driver.get(f'https://m.eais.go.kr/mbi/mbi/ada08/MBIADA08L02')
    time.sleep(1)

    # 페이지 안에 민원이 카테고리 버튼으로 나누어져 있어, 
    # 카테고리 버튼을 눌러서 하나씩 진행해야 함.
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.PAGE_DOWN)
#__layout > div > div > div.container > div.tab_content > div > div.all-complaint > div.sorting-list.js-sorting > div:nth-child(2) > button
    sorting_list = driver.find_elements(By.CSS_SELECTOR, 'div.sorting-list.js-sorting > div')
    for sorting in sorting_list:
        try:
            # display: none이 아닌 버튼만 클릭
            button = sorting.find_element(By.TAG_NAME, 'button')
            if button.is_displayed():
                button.click()
                time.sleep(2)

                divs = driver.find_elements(By.CSS_SELECTOR, 'div.all-complaint-list.base-side > div > div> button')
                time.sleep(3)
                print(len(divs))

        except Exception as e:
            print(f"Button not found or not clickable in div: {sorting}, Error: {str(e)}")

category()
```

```python
# 일단 카테고리별로 선택해서 들어가야 한다고 생각 했으나, 안에 데이터가 display=none으로 되어 있음
# 그러므로 그냥 한 페이지에서 다 뽑아낼 수 있음.

from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.by import By

import time
from selenium.webdriver.common.keys import Keys
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait

import pandas as pd

def crawling():

    driver = webdriver.Chrome()
    driver.get(f'https://m.eais.go.kr/mbi/mbi/ada08/MBIADA08L02')
    time.sleep(1.5)

    # 페이지 안에 민원이 카테고리 버튼으로 나누어져 있어, 
    # 카테고리 버튼을 눌러서 하나씩 진행해야 함.
    # body = driver.find_element(By.TAG_NAME, 'body')
    # body.send_keys(Keys.PAGE_DOWN)

    buttons = driver.find_elements(By.CSS_SELECTOR, 'div.all-complaint-list.base-side > div > div> button')
    time.sleep(1.5)

    complaints_list = []

# __layout > div > div:nth-child(1) > div > div.complaint-tbl-area > div > table
      
    for i in range(len(buttons)):
        try:  
            buttons = driver.find_elements(By.CSS_SELECTOR, 'div.all-complaint-list.base-side > div > div> button')

            driver.execute_script("arguments[0].scrollIntoView();", buttons[i])

            time.sleep(1.5)

            buttons[i].click()
            driver.implicitly_wait(2) 
            table_dict = {}
            complaint_table_area = driver.find_element(By.CLASS_NAME, 'complaint-tbl-area')
            complaint_table = complaint_table_area.find_element(By.TAG_NAME, 'table')
            rows = complaint_table.find_elements(By.CSS_SELECTOR, 'tbody tr')

            for row in rows:
                time.sleep(3)
                # 첫 번째 컬럼 (th)
                first_column = row.find_element(By.CSS_SELECTOR, 'th').text
                
                # 두 번째 컬럼 (td)
                second_column = row.find_element(By.CSS_SELECTOR, 'td')
                
                # 두 번째 컬럼 안의 모든 텍스트를 가져옴
                second_column_text = second_column.text.strip()

                
                # 딕셔너리에 값 추가
                table_dict[first_column] = second_column_text
            
            complaints_list.append(table_dict)
            
            driver.back()
            time.sleep(1)

        except Exception as e:
            print(f"뭔가가 잘못됨")
            print(len(complaints_list))
            df = pd.Dataframe(complaints_list)
            csv_file_path = "results/results.csv"
            df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
            print("저장 완")

    df = pd.Dataframe(complaints_list)
    csv_file_path = "results/results.csv"
    df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
    print("저장 완")
    
    driver.quit()

```
