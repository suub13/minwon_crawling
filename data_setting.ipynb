### 정리한 데이터 가져오기


```python
import pandas as pd
file_path = 'results/gov_faq_data.csv'
df = pd.read_csv(file_path, index_col=0)
```

```python
complaints_df = df[df['category'].str.contains('민원발급')]
```

#### 1. question: question value, answer: answer value

```python
new_dict = {}
complaints_list = []

for _, row in complaints_df.iterrows():
    complaints_list.append({
        "question": row['question'],
        "answer": row['answer']
    })

new_dict = {"민원발급":complaints_list}
```

```python
print(new_dict)
print(len(new_dict['민원발급']))
```

```python
import json

file_path = "results/complaints.json"

with open(file_path, 'w', encoding="utf-8") as json_file:
    json.dump(new_dict, json_file, ensure_ascii=False, indent=4)
```

#### 2. question value: answer value

```python
print(complaints_df)
```

```python
complaints_list2 = []

for _, row in complaints_df.iterrows():
    complaints_list2.append({
        row['question'] : row['answer']
    })

new_dict2 = {"민원발급":complaints_list2}
```

```python
import json

file_path = "results/complaints2.json"

with open(file_path, 'w', encoding="utf-8") as json_file:
    json.dump(new_dict2, json_file, ensure_ascii=False, indent=4)
```

#### 3.question: answer JSONL 형식 

```python
complaints_list3 = []

for _, row in complaints_df.iterrows():
    complaints_list3.append({
        row['question'] : row['answer']
    })
```

```python
import json

file_path = "results/complaints3.jsonl"

with open(file_path, 'w', encoding="utf-8") as jsonl_file:
    for complaint in complaints_list3:
        jsonl_file.write(json.dumps(complaint, ensure_ascii=False) + '\n')
```

#### 4. question: question, answer:answer JSONL 형식

```python
complaints_list4 = []

for _, row in complaints_df.iterrows():
    complaints_list4.append({
        "question":row['question'],
        "answer": row['answer']
    })
```

```python
import json

file_path = "results/complaints4.jsonl"

with open(file_path, 'w', encoding="utf-8") as jsonl_file:
    for complaint in complaints_list4:
        jsonl_file.write(json.dumps(complaint, ensure_ascii=False) + '\n')
```
