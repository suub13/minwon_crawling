```python
# string을 split해서, list의 첫 번째 내용만 골라서 그 내용이 있는 체크박스만 선택하고 타운로드 하기

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()

driver.get('https://www.law.go.kr/lsSc.do?menuId=1&subMenuId=15&query=고등교육법&dt=20201211#undefined')  # 원하는 URL로 변경

for _, key in enumerate(law_dict):
    print(key)
    # 검색 창 
    input_element = driver.find_element(By.XPATH, '/html/body/div[2]/div[1]/div[3]/div/div/input')

    # 값 입력
    input_element.clear()  # 기존 값을 지움
    input_element.send_keys(key)  # 원하는 값을 입력

    time.sleep(0.5)

    # 검색 버튼
    search_button = driver.find_element(By.XPATH, '//*[@id="sr_area"]/div/button')
    search_button.click()

    time.sleep(0.5)

    # 저장 버튼
    save_button = driver.find_element(By.XPATH, '//*[@id="bdySaveBtn"]/img')
    save_button.click()

    time.sleep(5)

    # 저장하는 내용 선택하기
    for value in law_dict[key]:
        laws_splits = value.split()
        print(laws_splits[0])
        xpath = f'//label[contains(text(), "{laws_splits[0]}")]'
        # 가 포함된 label을 찾아 해당 체크박스를 체크
        labels = driver.find_elements(By.XPATH, xpath)

        for label in labels:
            # label의 바로 앞에 있는 checkbox를 선택
            checkbox = label.find_element(By.XPATH, './preceding-sibling::input[@type="checkbox"]')
            if not checkbox.is_selected():
                checkbox.click()
        time.sleep(5)
    time.sleep(10)


```
