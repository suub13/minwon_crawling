엑셀파일 데이터 프레임으로 읽어오기

```python
import pandas as pd
import re

# 엑셀 파일 불러오기
file_path = "C:/Users/SOLI/Desktop/AI프로젝트 데이터/정부24/정부24데이터.xlsx"  # 엑셀 파일 경로
df = pd.read_excel(file_path, sheet_name='필터링데이터_38개')

# '근거법령' 열에서 앞의 숫자와 마침표를 제거하고 필요한 부분만 추출
def extract_law(text):
    # \n으로 구분된 법령들을 분리하여 리스트로 변환
    law_list = text.split('\n')
    # 각 항목의 앞에 숫자와 마침표 제거
    law_list = [re.sub(r'^\d+\.\s*', '', law).strip() for law in law_list]
    return law_list

# '근거법령' 열에서 법령 정보 추출
df['정리된근거법령'] = df['근거법령'].apply(extract_law)

# 모든 행의 법령 리스트를 하나로 합치기
all_laws = sum(df['정리된근거법령'], [])

# 전체 중복된 값 제거
unique_laws = list(set(all_laws))

print(unique_laws)

unique_laws_sorted = sorted(unique_laws)

len(unique_laws_sorted)
```

# 법: 조항 딕셔너리 정리

```python
import re
from collections import defaultdict

# 결과를 저장할 딕셔너리 생성
laws_dict = defaultdict(list)

# 리스트 처리
for law in unique_laws:
    # 괄호를 기준으로 분리
    law_name, law_clause = re.match(r"(.+?)\((.*)\)", law).groups()
    law_clause = law_clause.strip()  # 괄호 안의 내용을 정리

    # 쉼표로 구분된 조항들을 분리
    clauses = [clause.strip() for clause in law_clause.split(',')]
    
    # 괄호 안이 비어있다면 'all' 추가, 그렇지 않으면 조항 추가
    if not law_clause:
        laws_dict[law_name].append('all')
    else:
        laws_dict[law_name].extend(clauses)

sorted_dict = {key: laws_dict[key] for key in sorted(laws_dict)}

# 결과 출력
for key, value in sorted_dict.items():
    sorted_dict[key] = sorted_dict[key].sort()
    print(f"'{key}': {value}")

```

### 위에 내용과 다르게 법령 (조 제외) 기준

```python
import pandas as pd
import re

# 엑셀 파일 불러오기
file_path = '민원처리기준표_0905.xlsx'  # 엑셀 파일 경로
df = pd.read_excel(file_path, sheet_name='필터링데이터')

# '근거법령' 열에서 괄호 앞 부분만 추출하는 함수
def extract_law_without_parenthesis(text):
    # \n으로 구분된 법령들을 분리하여 리스트로 변환
    law_list = text.split('\n')
    # 괄호 앞 부분만 추출
    law_list = [re.sub(r'^\d+\.\s*|\s*\(.*\)', '', law).strip() for law in law_list]
    return law_list

# '근거법령' 열을 처리하여 괄호 앞 부분의 법령 리스트 만들기
df['정리된근거법령리스트'] = df['근거법령'].apply(extract_law_without_parenthesis)

# 모든 행의 법령 리스트를 하나로 합치기
all_laws = sum(df['정리된근거법령리스트'], [])

# 전체 중복된 값 제거
unique_laws = list(set(all_laws))

# 고유한 법령 리스트 출력
unique_laws.sort()  # 보기 좋게 정렬
```
